```python
import pandas as pd
import numpy as np
import os
import matplotlib.pylab as plt
from pandas.plotting import register_matplotlib_converters
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15,9
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.tsa.arima_model import ARIMA
```

```python
file1 = 'US Coffee C Futures Historical Data.csv'
table1 = pd.read_csv(file1)
table1
```

```python
file2 = 'US Coffee C Futures Historical Data2.csv'
table2 = pd.read_csv(file2)
table2
```

```python
file3 = 'US Coffee C Futures Historical Data3.csv'
table3 = pd.read_csv(file3)
table3
```

```python
data_frames = [table1,table2,table3]
df = pd.concat(data_frames)
df = df.drop_duplicates()
df
```

```python
df['Date']=df['Date'].astype(str)
df['Date']=pd.to_datetime(df['Date'])
df = df.set_index(['Date'])
```

```python
df
```

```python
df.sort_index(inplace=True)
df
```

```python
# Select only price

df = df.loc[:,['Price']]
```

```python
# Plot graph first

# plot graph

plt.xlabel("Date")
plt.ylabel("Price")
plt.plot(df)
register_matplotlib_converters()
```

```python
# Determining roling statistics
rolmean = df.rolling(window=12).mean()
rolstd = df.rolling(window=12).std()
print(rolmean,rolstd)
```

```python
# Plot rolling statistics

orig = plt.plot(df, color='blue',label='Original')
mean = plt.plot(rolmean, color='red',label='Rolling Mean')
std = plt.plot(rolstd, color='yellow',label='Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviaion')
plt.show(block=False)
```

```python
# Perform Dickey-Fuller test:

print("Result of Dickey-Fuller Test:")
dftest=adfuller(df['Price'],autolag='AIC')

dfoutput = pd.Series(dftest[0:4], index=['Test Statistics','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critial Value (%s)'%key]=value
    
print(dfoutput)
```

```python
# Estimating trend
df_logScale = np.log(df)
plt.plot(df_logScale)
```

```python
# Determining roling statistics
rolmean = df_logScale.rolling(window=52).mean()
rolstd = df_logScale.rolling(window=52).std()
print(rolmean,rolstd)
```

```python
# Plot rolling statistics

orig = plt.plot(df_logScale, color='blue',label='Original')
mean = plt.plot(rolmean, color='red',label='Rolling Mean')
std = plt.plot(rolstd, color='yellow',label='Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviaion')
plt.show(block=False)
```

```python
df_logScale_minus_movAverage = df_logScale-rolmean
df_logScale_minus_movAverage
```

```python
# Remove nan values

df_logScale_minus_movAverage.dropna(inplace=True)
df_logScale_minus_movAverage.head(10)
```

```python
def test_stationary(timeseries):
    
    # Determining rolling stats
    movingAverage = timeseries.rolling(window=52).mean()
    movingStd = timeseries.rolling(window=52).std()
    
    # Plot rolling stats
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(movingAverage, color='red',label='Rolling Mean')
    std = plt.plot(movingStd, color='yellow',label='Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviaion')
    plt.show(block=False)
    
    # Perform Dickey-Fuller test:

    print("Result of Dickey-Fuller Test:")
    dftest=adfuller(timeseries['Price'],autolag='AIC')

    dfoutput = pd.Series(dftest[0:4], index=['Test Statistics','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critial Value (%s)'%key]=value

    print(dfoutput)
```

```python
test_stationary(df_logScale_minus_movAverage)
```

```python
exponentialDecayWeightedAverage = df_logScale.ewm(halflife=52,min_periods=0,adjust=True).mean()
plt.plot(df_logScale)
plt.plot(exponentialDecayWeightedAverage, color='red')
```

```python
df_logScale_exponentialDecayWeightedAverage = df_logScale-exponentialDecayWeightedAverage
test_stationary(df_logScale_exponentialDecayWeightedAverage)
```

```python
df_logScale_shifted = df_logScale-df_logScale.shift()
plt.plot(df_logScale_shifted)
```

```python
decomposition = seasonal_decompose(df_logScale,freq=30)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(411)
plt.plot(df_logScale, label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal, label='Seasonal')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residual')
plt.legend(loc='best')

decomposedLogData = residual
decomposedLogData.dropna(inplace=True)
test_stationary(decomposedLogData)
```

```python
# ACF and PACF plots
df_logScale_minus_movAverage_shifted=df_logScale_minus_movAverage-df_logScale_minus_movAverage.shift()
df_logScale_minus_movAverage_shifted.replace([np.inf, -np.inf], np.nan).dropna(inplace=True)
lag_acf = acf(df_logScale_minus_movAverage_shifted, nlags=20)
lag_pacf = pacf(df_logScale_minus_movAverage_shifted,nlags=20)

# Plot ACF
plt.subplot(121)
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(df_logScale_minus_movAverage_shifted)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(df_logScale_minus_movAverage_shifted)),linestyle='--',color='gray')
plt.title('Autocorrelation function')

# Plot PACF
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(df_logScale_minus_movAverage_shifted)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(df_logScale_minus_movAverage_shifted)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation function')
plt.tight_layout()
```

```python
df_logScale.index = pd.DatetimeIndex(df_logScale.index.values,freq=df_logScale.index.inferred_freq)
```

```python
# AR Model
model = ARIMA(df_logScale,order=(2,1,2))
results_AR = model.fit(disp=-1)
plt.plot(df_logScale_shifted)
plt.plot(results_AR.fittedvalues,color='red')
plt.title('RSS: %.4f'% sum(results_AR.fittedvalues-df_logScale_shifted['Price'])**2)
print("Plotting AR-Model")
```
